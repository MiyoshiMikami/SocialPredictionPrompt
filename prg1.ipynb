## PROMPT COMPILER
In this project you will see how transform a natural text of instrucction into the formal prompt with properties:
    ** Temperature
    ** Max Length
    ** Variables

```python
from openai import openAI
client = openAI

print("Ingresa la proposici√≥n a demostrar")
proposition = input()


response = client.response.create(
    model = "gpt-4",
    reasoing = {"effort": "low"},
    temperature = "cold",
    input = [
            {
                "role"="user",
                "content"= proposition
            }
    ]
)

print(response.output_text)
```
