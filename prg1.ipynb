## PROMPT COMPILER
In this project you will see how transform a natural text of instrucction into the formal prompt with properties:
    ** Temperature
    ** Max Length
    ** Variables

```python
# Create and import the libs and create the API connection
from openai import OpenAI
client = OpenAI()

# Input for a user to create the response
print("Ingresa la proposici√≥n a demostrar")
proposition = input() 

# Correct API call
response = client.chat.completions.create(
    model="gpt-4",
    messages=[
        {
            "role": "user",
            "content": proposition
        }
    ],
    temperature=0.7,  # 0.0 to 2.0, not "cold"
    max_tokens=500  # Optional: limit response length
)

# Print the response
print(response.choices[0].message.content)
```
